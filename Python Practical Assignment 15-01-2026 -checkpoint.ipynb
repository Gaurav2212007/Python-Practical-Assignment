should return the index of element to be searched in the list.

```python
# Sequential search in a sorted list

def sequential_search(lst, item):
    for i in range(len(lst)):
        if lst[i] == item:
            return i   # return index
    return -1          # not found

# Example
numbers = [10, 20, 30, 40, 50]
search_item = int(input("Enter number to search: "))

pos = sequential_search(numbers, search_item)

if pos != -1:
    print("Item found at index:", pos)
else:
    print("Item not found")
```

```python
# Program to find maximum and minimum in a tuple

numbers = (5, 2, 8, 1, 9, 3)

# initialize both with first element
maximum = numbers[0]
minimum = numbers[0]

# Checking each element in the tuple
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

#Printing the Results
print("Maximum:", maximum)
print("Minimum:", minimum)
```

```python
## Program to find maximum and minimum in a tuple without using max()

# take tuple input from user
n = int(input("Enter how many elements: "))
elements = []

for i in range(n):
    num = int(input("Enter element: "))
    elements.append(num)

numbers = tuple(elements)   # convert list to tuple

# initialize max and min with first element
maximum = numbers[0]
minimum = numbers[0]

for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("Tuple:", numbers)
print("Maximum:", maximum)
print("Minimum:", minimum)
```

```python
# Program: Count occurrence of each element in a tuple (without repeating output)

n = int(input("Enter how many elements: "))
elements = []
for i in range(n):
    num = int(input("Enter element: "))
    elements.append(num)

tup = tuple(elements)

# list to remember which elements are already counted
done = []

for i in range(len(tup)):
    item = tup[i]
    if item not in done:          # only count if not already done
        count = 0
        for j in range(len(tup)):
            if tup[j] == item:
                count = count + 1
        print(item, ":", count)
        done.append(item)         # mark this element as counted
```

```python
# Program: Find second largest element in a tuple

n = int(input("Enter how many elements: "))
elements = []
for i in range(n):
    num = int(input("Enter element: "))
    elements.append(num)

tup = tuple(elements)

if len(tup) < 2:
    print("Second largest does not exist")
else:
    # Initialize
    largest = tup[0]
    second = None

    # First pass: find largest
    for x in tup:
        if x > largest:
            largest = x

    # Second pass: find second largest (less than largest)
    for x in tup:
        if x != largest:
            if second is None or x > second:
                second = x

    if second is None:
        print("Second largest does not exist (all elements equal)")
    else:
        print("Tuple:", tup)
        print("Largest:", largest)
        print("Second largest:", second)
```

<h4>Q2.</h4>Write a program to sort a tuple of tuples based on the second element. 
Example: ((1, 5), (3, 2), (2, 8)) should become ((3, 2), (1, 5), (2, 8))

```python
# Program: Sort a tuple of tuples by second element (ascending)

n = int(input("Enter how many pairs: "))
pairs = []
for i in range(n):
    a = int(input("Enter first value: "))
    b = int(input("Enter second value: "))
    pairs.append((a, b))

# Selection sort by second element
for i in range(len(pairs)):
    min_index = i
    for j in range(i + 1, len(pairs)):
        if pairs[j][1] < pairs[min_index][1]:
            min_index = j
    # swap
    temp = pairs[i]
    pairs[i] = pairs[min_index]
    pairs[min_index] = temp

sorted_tuple = tuple(pairs)

print("Sorted tuple:", sorted_tuple)
```

<h4>Q3.</h4>Write a function to calculate the factorial of a number (both iterative and recursive).

```python
def factorial_iterative(n):
    fact = 1
    for i in range(1, n+1):
        fact *= i
    return fact

# Function to calculate factorial recursively
def factorial_recursive(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial_recursive(n-1)

# Input number from user
num = int(input("Enter a number: "))

# Call iterative function
print("Factorial (iterative):", factorial_iterative(num))

# Call recursive function
print("Factorial (recursive):", factorial_recursive(num))
```

```python
class Rectangle:

    # Constructor to initialize values
    def __init__(self, length, width):
        self.length = length
        self.width = width

    # Method to calculate area
    def area(self):
        return self.length * self.width

    # Method to calculate perimeter
    def perimeter(self):
        return 2 * (self.length + self.width)

# Input from user
l = int(input("Enter length: "))
w = int(input("Enter width: "))

# Create object of class
r = Rectangle(l, w)

# Display results
print("Area of Rectangle:", r.area())
print("Perimeter of Rectangle:", r.perimeter())
```

<h4>Q5.</h4> Create a class Circle with method to calculate area. Overload the + operator to add two 
circles (add their radii).

```python
import math

class Circle:
    def __init__(self, radius):
        self.__radius = radius   # private attribute

    # Method to calculate area
    def area(self):
        return math.pi * (self.__radius ** 2)

    # Getter for radius
    def get_radius(self):
        return self.__radius

    # Operator overloading for +
    def __add__(self, other):
        # Add radii of two circles
        new_radius = self.__radius + other.__radius
        return Circle(new_radius)


# --- Input from user ---
r1 = float(input("Enter radius of first circle: "))
r2 = float(input("Enter radius of second circle: "))

# Create objects
c1 = Circle(r1)
c2 = Circle(r2)

# Display details
print("\nCircle 1 radius:", c1.get_radius())
print("Circle 1 area:", c1.area())

print("\nCircle 2 radius:", c2.get_radius())
print("Circle 2 area:", c2.area())

# Add two circles
c3 = c1 + c2
print("\nAfter adding circles:")
print("New Circle radius:", c3.get_radius())
print("New Circle area:", c3.area())

```

<h4>Q1.</h4>Create a class hierarchy: Animal -> Mammal -> Dog. Demonstrate multilevel inheritance. 

```python
# Program: Multilevel Inheritance (Animal -> Mammal -> Dog)

class Animal:
    def __init__(self, species):
        self.species = species

    def show_species(self):
        print("Species:", self.species)

    def sound(self):
        print("Animals make different sounds")


# Mammal inherits from Animal
class Mammal(Animal):
    def __init__(self, species, has_hair=True):
        super().__init__(species)   # call Animal constructor
        self.has_hair = has_hair

    def show_characteristics(self):
        print("Mammals usually have hair:", self.has_hair)

    def sound(self):
        print("Mammals make warm-blooded sounds")


# Dog inherits from Mammal
class Dog(Mammal):
    def __init__(self, breed):
        super().__init__("Dog")   # call Mammal constructor with species = Dog
        self.breed = breed

    def show_breed(self):
        print("Breed:", self.breed)

    def sound(self):
        print("Dog barks: Woof! Woof!")


# --- Example usage ---
d = Dog("German Shepherd")

# Methods from Animal
d.show_species()

# Methods from Mammal
d.show_characteristics()

# Methods from Dog
d.show_breed()

# Demonstrate overridden sound method
d.sound()
```

<h4>Q2.</h4>Create a class Employee with class method to keep count of total employees.

```python
# Program: Employee class with class method to count employees

class Employee:
    # class variable to keep count
    __count = 0

    def __init__(self, name, emp_id):
        self.name = name
        self.emp_id = emp_id
        # increment count whenever a new object is created
        Employee.__count += 1

    # class method to access count
    @classmethod
    def total_employees(cls):
        return cls.__count


# --- Example usage ---
e1 = Employee("Haathi", 101)
e2 = Employee("Ghoda", 102)
e3 = Employee("Palki", 103)

print("Total employees:", Employee.total_employees())
```

<h4>Q3.</h4>Create a class Book with special methods __str__, __len__, and __eq__. 

```python
# Program: Book class with special methods

class Book:

    def __init__(self, title, pages):
        self.title = title
        self.pages = pages

    # String representation
    def __str__(self):
        return "Book Title: " + self.title

    # Length of book (number of pages)
    def __len__(self):
        return self.pages

    # Compare two books
    def __eq__(self, other):
        return self.pages == other.pages

# Create objects
b1 = Book("Python", 200)
b2 = Book("Java", 200)

# Use special methods
print(b1)                # calls __str__
print("Pages:", len(b1)) # calls __len__
print("Books equal:", b1 == b2)  # calls __eq__
```

<h4>Q4.</h4>Create a class Temperature with property decorators for Celsius and Fahrenheit 
conversion. 

```python
# Program: Temperature class with property decorators

class Temperature:
    def __init__(self, celsius=0):
        self.__celsius = celsius   # private attribute

    # Celsius property
    @property
    def celsius(self):
        return self.__celsius

    @celsius.setter
    def celsius(self, value):
        self.__celsius = value

    # Fahrenheit property (calculated from Celsius)
    @property
    def fahrenheit(self):
        return (self.__celsius * 9/5) + 32

    @fahrenheit.setter
    def fahrenheit(self, value):
        # convert Fahrenheit to Celsius
        self.__celsius = (value - 32) * 5/9


# --- Example usage ---
t = Temperature()

# Set Celsius
t.celsius = float(input("Enter temperature in Celsius: "))
print("Celsius:", t.celsius)
print("Fahrenheit:", t.fahrenheit)

# Set Fahrenheit
t.fahrenheit = float(input("\nEnter temperature in Fahrenheit: "))
print("Celsius:", t.celsius)
print("Fahrenheit:", t.fahrenheit)
```

<h4>Q1.</h4>Create an abstract class Shape with abstract method area(). Create subclasses Rectangle, 
Circle, and Triangle that implement the area method.

```python
from abc import ABC, abstractmethod
import math

# Abstract class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass   # abstract method, must be implemented in subclasses


# Rectangle subclass
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width


# Circle subclass
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * (self.radius ** 2)


# Triangle subclass
class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height


# --- Example usage ---
# Rectangle
l = float(input("Enter length of rectangle: "))
w = float(input("Enter width of rectangle: "))
rect = Rectangle(l, w)
print("Area of Rectangle:", rect.area())

# Circle
r = float(input("\nEnter radius of circle: "))
circle = Circle(r)
print("Area of Circle:", circle.area())

# Triangle
b = float(input("\nEnter base of triangle: "))
h = float(input("Enter height of triangle: "))
tri = Triangle(b, h)
print("Area of Triangle:", tri.area())
```

```python
from abc import ABC, abstractmethod
# Abstract class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

# Rectangle class
class Rectangle(Shape):
    def area(self):
        return 10 * 5

# Circle class
class Circle(Shape):
    def area(self):
        return 3.14 * 7 * 7

# Triangle class
class Triangle(Shape):
    def area(self):
        return 0.5 * 8 * 6

# Objects
r = Rectangle()
c = Circle()
t = Triangle()

# Display area
print("Rectangle Area:", r.area())
print("Circle Area:", c.area())
print("Triangle Area:", t.area())
```

<h4>Q2.</h4>Create a class Library that manages a collection of books. Implement methods to add 
books, remove books, search books, and display all books.

```python
class Library:
    def __init__(self):
        self.books = []

    def add(self, book):
        self.books.append(book)

    def remove(self, book):
        self.books.remove(book)

    def search(self, book):
        print(book in self.books)

    def show(self):
        print(self.books)

# Object
l = Library()

l.add("Python")
l.add("Java")
l.add("C")

l.search("Python")
l.remove("Java")
l.show()
```

```python
# Program: Library class to manage books

class Library:
    def __init__(self):
        self.books = []   # collection of books stored in a list

    # Method to add a book
    def add_book(self, title):
        self.books.append(title)
        print(f"Book '{title}' added to library.")

    # Method to remove a book
    def remove_book(self, title):
        if title in self.books:
            self.books.remove(title)
            print(f"Book '{title}' removed from library.")
        else:
            print(f"Book '{title}' not found in library.")

    # Method to search for a book
    def search_book(self, title):
        if title in self.books:
            print(f"Book '{title}' is available in library.")
        else:
            print(f"Book '{title}' is NOT available in library.")

    # Method to display all books
    def display_books(self):
        if len(self.books) == 0:
            print("Library is empty.")
        else:
            print("Books in library:")
            for book in self.books:
                print("-", book)


# --- Example usage ---
lib = Library()

# Add books
lib.add_book("Python Programming")
lib.add_book("Data Structures")
lib.add_book("Machine Learning")

# Display books
lib.display_books()

# Search for a book
lib.search_book("Python Programming")
lib.search_book("Java")

# Remove a book
lib.remove_book("Data Structures")

# Display after removal
lib.display_books()
```
